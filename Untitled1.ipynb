```python
import numpy as np
import math
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
ms_2_00 = []
ms_2_15 = []
ms_2_30 = []
ms_5_00 = []
ms_5_30 = []
mw_2_00 = []
mw_2_30 = []
path = '~/Desktop/experiment_csv/'
```

```python
ms_2_00 = pd.read_csv(path+'ms_2_00.csv',header=None)
ms_2_15 = pd.read_csv(path+'ms_2_15.csv',header=None)
ms_2_30 = pd.read_csv(path+'ms_2_30.csv',header=None)
ms_5_00 = pd.read_csv(path+'ms_5_00.csv',header=None)
ms_5_30 = pd.read_csv(path+'ms_5_30.csv',header=None)
mw_2_00 = pd.read_csv(path+'mw_2_00.csv',header=None)
mw_2_30 = pd.read_csv(path+'mw_2_30.csv',header=None)
max_ms_2 = [(664.000000,0.745883), (664.000000,0.741474),(658.000000,0.182685), ]
max_ms_5 = [(664.000000,0.777184), (659.000000,0.263588)]
max_mw_2 = [(664.000000,0.773358), (664.000000,0.414299)]
```

```python
idx = np.array(ms_5_30.iloc[:350,:2]).argmax(axis=0).tolist()
ms_5_30.iloc[idx[1]]
```

```python
fig, axes = plt.subplots(1,3,figsize=(15,5))
for ax,df,peak in zip(axes, [ms_2_00, ms_2_30, ms_2_15], max_ms_2):
    ax.scatter(df[0],df[1])
    ax.plot(peak[0],peak[1],marker='x',c='r', markersize="15")
    ax.text(peak[0]-50,peak[1]+1, (round(peak[0],2), round(peak[1],2)),fontsize="15")
    ax.set_xlabel('波長 nm', fontsize="15")
    ax.set_ylabel('吸光度',fontsize="15")
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(20,5))
for ax,df,peak in zip(axes, [ms_5_00, ms_5_30], max_ms_5):
    ax.scatter(df[0],df[1])
    ax.plot(peak[0],peak[1],marker='x',c='r', markersize="20")
    ax.text(peak[0]-50,peak[1]+1, (round(peak[0],2), round(peak[1],2)),fontsize="20")
    ax.set_xlabel('波長 nm', fontsize="20")
    ax.set_ylabel('吸光度',fontsize="20")
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(20,5))
for ax,df,peak in zip(axes, [mw_2_00, mw_2_30], max_mw_2):
    ax.scatter(df[0],df[1])
    ax.plot(peak[0],peak[1],marker='x',c='r', markersize="20")
    ax.text(peak[0]-50,peak[1]+0.1, (round(peak[0],2), round(peak[1],2)),fontsize="20")
    ax.set_xlabel('波長 nm', fontsize="20")
    ax.set_ylabel('吸光度',fontsize="20")
fig.tight_layout()
plt.show()
```
